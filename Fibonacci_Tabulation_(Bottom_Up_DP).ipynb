```
def fibonacci_tabulation(n: int) -> int:
    """
    Calculates the nth Fibonacci number using Tabulation (Bottom-Up DP).
    Time complexity: O(n) - Iterative approach is often faster in practice.
    """
    if n <= 1:
        return n

    # 1. Create a DP table (array)
    # The table stores the solution to the subproblem F(i) at index i.
    dp = [0] * (n + 1)

    # 2. Initialize base cases
    dp[0] = 0
    dp[1] = 1

    # 3. Fill the table iteratively (from smallest subproblem to largest)
    for i in range(2, n + 1):
        # The recurrence relation: dp[i] depends on dp[i-1] and dp[i-2]
        dp[i] = dp[i - 1] + dp[i - 2]

    # 4. The result is the final element in the table
    return dp[n]

# print(f"F(6) Tabulated: {fibonacci_tabulation(6)}") # Output: 8
# print(f"F(50) Tabulated: {fibonacci_tabulation(50)}") # Runs instantly
```
